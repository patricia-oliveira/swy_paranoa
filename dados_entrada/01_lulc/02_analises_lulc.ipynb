```python
import pandas as pd
import geopandas as gpd
```

```python
uh = gpd.read_file("../bh_lago_paranoa.geojson").to_crs(4674)
display(uh)
ra = gpd.read_file("../bh_ra.geojson").to_crs(4674)
display(ra)
```

```python
dn_to_name = {
    3: "Formação Florestal", 4: "Formação Savânica", 5: "Mangue", 6: "Floresta Alagável", 9: "Silvicultura", 11: "Campo Alagado e Área Pantanosa",
    12: "Formação Campestre", 15: "Pastagem", 19: "Lavoura Temporária",  20: "Cana-de-Açúcar", 21: "Mosaico de Usos", 24: "Praia, Duna e Areal",
    24: "Área Urbanizada", 25: "Outras Áreas Não Vegetadas", 27: "Não Observado", 29: "Afloramento Rochoso", 30: "Mineração", 31: "Aquicultura",
    33: "Rio, Lago e Oceano", 35: "Dendê", 39: "Soja", 40: "Arroz", 41: "Outras Lavouras Temporárias", 46: "Café", 47: "Citrus", 
    48: "Outras Lavouras Perenes", 62: "Algodão"
}
dn_to_name_simpl = {
    3: "Floresta", 4: "Floresta", 5: "Floresta", 6: "Floresta", 9: "Silvicultura", 11: "Vegetação Herbácea e Arbustiva", 
    12: "Vegetação Herbácea e Arbustiva", 15: "Pastagem", 19: "Agricultura",  20: "Agricultura", 21: "Agricultura", 24: "Praia, Duna e Areal",
    24: "Área Urbanizada", 25: "Outras Áreas Não Vegetadas", 27: "Não Observado", 29: "Vegetação Herbácea e Arbustiva", 30: "Mineração",
    31: "Aquicultura", 33: "Rio, Lago e Oceano", 35: "Agricultura", 39: "Agricultura", 40: "Agricultura", 41: "Agricultura", 46: "Agricultura",
    47: "Agricultura", 48: "Agricultura", 62: "Agricultura"
}
```

```python
lulc_85 = gpd.read_file(f"./lulc_data/land_use_30m_1985.geojson")
#lulc_85['description'] = lulc_85['lucode'].map(dn_to_name)
lulc_85['description_simpl'] = lulc_85['lucode'].map(dn_to_name_simpl)
lulc_85 = lulc_85[["description_simpl", "geometry"]].dissolve(by = ["description_simpl"]).reset_index()
```

```python
lulc_24 = gpd.read_file(f"./lulc_data/land_use_30m_2024.geojson")
#lulc_24['description'] = lulc_24['lucode'].map(dn_to_name)
lulc_24['description_simpl'] = lulc_24['lucode'].map(dn_to_name_simpl)
lulc_24 = lulc_24[["description_simpl", "geometry"]].dissolve(by = ["description_simpl"]).reset_index()
```

```python
def area_porcentagem(gdf):
    gdf = gdf.to_crs('EPSG:31983')  
    gdf['area'] = gdf.geometry.area
    gdf['area_ha'] = gdf['area'] / 10000  
    gdf['porc_area'] = (gdf['area_ha'] / gdf['area_ha'].sum()) * 100
    gdf = gdf.to_crs('EPSG:4674')
    return gdf
```

```python
lulc_85 = area_porcentagem(lulc_85)
lulc_24 = area_porcentagem(lulc_24)
```

```python
lulc_24 = gpd.read_file(f"./lulc_data/land_use_30m_2024.geojson")
#lulc_24['description'] = lulc_24['lucode'].map(dn_to_name)
lulc_24['description'] = lulc_24['lucode'].map(dn_to_name)
lulc_24 = lulc_24.dissolve(by = "description").reset_index()
lulc_24 = area_porcentagem(lulc_24)
```

```python
lulc_24[~lulc_24.description.isna()].sort_values("porc_area", ascending = False)
```

```python
pd.merge(lulc_85[["description_simpl", "porc_area"]], 
         lulc_24[["description_simpl", "porc_area"]], 
         on = "description_simpl", suffixes=('_85', '_24')).sort_values("porc_area_85", ascending=False)
```

```python
uh_lulc_85 = gpd.overlay(uh, lulc_85, how = "intersection").dissolve(by = ["uh_label", "description_simpl"]).reset_index()
uh_lulc_85 = area_porcentagem(uh_lulc_85)
uh_lulc_24 = gpd.overlay(uh, lulc_24, how = "intersection").dissolve(by = ["uh_label", "description_simpl"]).reset_index()
uh_lulc_24 = area_porcentagem(uh_lulc_24)
uh_lulc = pd.merge(uh_lulc_85[["uh_label", "description_simpl", "porc_area"]], 
                   uh_lulc_24[["uh_label", "description_simpl", "porc_area"]], 
                   on = ["uh_label", "description_simpl"], suffixes=('_85', '_24'))
uh_lulc["dif_porc_area"] = uh_lulc["porc_area_24"] - uh_lulc["porc_area_85"]
```

```python
uh_lulc.sort_values("dif_porc_area", ascending=False)
```

```python
ra_lulc_85 = gpd.overlay(ra, lulc_85, how = "intersection").dissolve(by = ["ra", "description_simpl"]).reset_index()
ra_lulc_85 = area_porcentagem(ra_lulc_85)
```

```python
display(uh_lulc_85[uh_lulc_85.description_simpl.isin(["Floresta", "Vegetação Herbácea e Arbustiva"])].sort_values("porc_area", ascending=False).head(5))
display(uh_lulc_24[uh_lulc_24.description_simpl.isin(["Floresta", "Vegetação Herbácea e Arbustiva"])].sort_values("porc_area", ascending=False).head(5))
```

```python
ra_lulc_85 = gpd.overlay(ra, lulc_85, how = "intersection").dissolve(by = ["ra", "description_simpl"]).reset_index()
ra_lulc_85 = area_porcentagem(ra_lulc_85)
ra_lulc_24 = gpd.overlay(ra, lulc_24, how = "intersection").dissolve(by = ["ra", "description_simpl"]).reset_index()
ra_lulc_24 = area_porcentagem(ra_lulc_24)
ra_lulc = pd.merge(ra_lulc_85[["ra", "description_simpl", "porc_area"]], 
         ra_lulc_24[["ra", "description_simpl", "porc_area"]], 
         on = ["ra", "description_simpl"], suffixes=('_85', '_24')).sort_values("porc_area_85", ascending=False)
ra_lulc["dif_porc_area"] = ra_lulc["porc_area_24"]  - ra_lulc["porc_area_85"] 
```

```python
ra_lulc[ra_lulc.description_simpl.isin(["Área Urbanizada"])].sort_values("dif_porc_area", ascending=False)
```
